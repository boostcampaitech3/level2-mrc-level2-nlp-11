```python
import matplotlib.pyplot as plt
```

```python
topk = range(1, 21)
validation_score = [30.00, 42.50, 48.75, 52.92, 56.67, 59.17, 60.42, 63.33, 64.17, 65.83, 66.67, 67.50, 69.58, 70.00, 72.08, 72.08, 72.92, 73.33, 73.75, 73.75]
```

```python
font_title = {
    'color':  'black',
    'weight': 'bold',
    'size': 14
}
font1 = {
    'color':  'black',
    'weight': 'normal',
    'size': 14
}
font2 = {
    'color':  'blue',
    'weight': 'bold',
    'size': 11
}

plt.figure(figsize=(12, 10))
plt.plot(topk, validation_score)

plt.title('TF-IDF 50000', fontdict=font_title)
plt.ylabel('Correct-%', fontdict=font1)
plt.xlabel('Top K', fontdict=font1)
plt.ylim([0, 100])
plt.xticks(range(1, 21))

plt.text(0.4, 27, '30.00%', fontdict=font2)
plt.text(19.4, 76, '73.75%', fontdict=font2)

# plt.show()
plt.savefig('TF-IDF_50000.jpg')
```
